```python
from rdflib import Graph
```

```python
%reload_ext jupyter-rdfify
```

```python
g = Graph()
g.parse("courseKG.ttl")
```

```python
s = g.serialize(format='ttl')
print(s)
```

```python
#What are the modules for MMDS students
sparql = """
    SELECT ?modules ?program
    WHERE {
        ?modules prop:isModuleOf  ?program .
        }
      """  
```

```python
for r in g.query(sparql):
    print(r)
    
```

```python
#What are the courses in Data Management for MMDS students
sparql = """
    SELECT ?courses
    WHERE {
        ?courses prop:isCourseOf  <https://www.uni-mannheim.de/en/academics/programs/mannheim-master-in-data-science/modules/Data_Management> .
        #FILTER(?module=="<https://www.uni-mannheim.de/en/academics/programs/mannheim-master-in-data-science/modules/Data_Management>")
        
        }
      """ 
```

```python
for r in g.query(sparql):
    print(r)
```
